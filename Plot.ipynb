```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl

# Set the default color cycle
mpl.rcParams['axes.prop_cycle'] = mpl.cycler(color=["r", "k", "c"]) 
```

```python
tweetsSentiment = ["Positivos", "Negativos"]
tweets = [61, 39]
x = np.arange(2)

fig, ax = plt.subplots()
barlist = plt.bar(x, tweets)
plt.xticks(x, tweetsSentiment)
barlist[0].set_color('g')
plt.title("Tweets Positivos e Negativos em Porcentagem- Boulos")
fig.set_figheight(7)
fig.set_figwidth(10)
plt.show()
```

```python
tweetsSentiment = ["Márcio França", "Arthur do Val", "Boulos", "Russomanno", "Covas"]
tweets = [66, 62, 61, 60, 59]
x = np.arange(5)

fig, ax = plt.subplots()
barlist = plt.barh(x, tweets, color=['green', 'cyan', 'blue', 'orange', 'red'])
plt.yticks(x, tweetsSentiment)
plt.title("Tweets Positivos em Porcentagem")
fig.set_figheight(7)
fig.set_figwidth(10)
plt.show()
```
